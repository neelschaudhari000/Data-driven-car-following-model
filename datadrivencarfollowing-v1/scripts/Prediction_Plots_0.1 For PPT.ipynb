```python
import FileProcessing
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly
import warnings
warnings.filterwarnings("ignore")
```

```python
file=FileProcessing.FileProcessing()
prediction_df_01 = file.read_input('Prediction_set_Predited_data_0.1')
def selectpair(pair):
    LF = prediction_df_01[prediction_df_01["L-F_Pair"] == pair]
    return LF
```

```python
prediction_df_01["Pair Time Duration"] = prediction_df_01["pair_Time_Duration"]
prediction_df_01["Actual Velocity"] = prediction_df_01["nextframesvel"]
prediction_df_01["Actual Acceleration"] = prediction_df_01["nextframeAcc"]
prediction_df_01["Actual Jerk"] = prediction_df_01["nextframejerk"]
prediction_df_01["Actual Spacing"] = prediction_df_01["nextFrameSpacing"]
```

```python
LF1 = selectpair('2322-2330')
```

```python
def plot(df,xlabel,ylabel,actual,cnnprediction,cnnlabel,title):
      #fig1, ax = plt.subplots()
      plt.figure(figsize=(15, 9))
      plt.title(title,color = "black", size = 18)
      plt.xlabel(xlabel, color = "black", size = 18)
      plt.ylabel(ylabel, color = "black", size = 18)
      plt.rc('xtick', labelsize=15)
      plt.rc('ytick', labelsize=15)
      plt.xticks( range(0,60,5) )
      sns.lineplot(x = df["Pair Time Duration"],y= df[actual],color="black" ,label = actual)
      ax=sns.lineplot(x = df["Pair Time Duration"],y= df[cnnprediction],color="blue" ,label = cnnlabel)
      ax.spines['right'].set_visible(False)
      ax.spines['top'].set_visible(False)
      ax.spines['left'].set_visible(False)
      ax.spines['bottom'].set_visible(False)
      ax.tick_params(left = False, bottom = False)
      return plt
def plot1(df,xlabel,ylabel,actual,cnnprediction,rfprediction,cnnlabel,rflabel,title):
      plt.figure(figsize=(15, 9))
      plt.title(title,color = "black", size = 18)
      plt.xlabel(xlabel, color = "black", size = 18)
      plt.ylabel(ylabel, color = "black", size = 18)
      plt.rc('xtick', labelsize=15)
      plt.rc('ytick', labelsize=15)
      plt.xticks( range(0,60,5) )
      sns.lineplot(x = df["Pair Time Duration"],y= df[actual],color="black" ,label = actual)
      ax=sns.lineplot(x = df["Pair Time Duration"],y= df[rfprediction],color="blue" ,label = rflabel)
      ax=sns.lineplot(x = df["Pair Time Duration"],y= df[cnnprediction],color="grey",label = cnnlabel)
      ax.spines['right'].set_visible(False)
      ax.spines['top'].set_visible(False)
      ax.spines['left'].set_visible(False)
      ax.spines['bottom'].set_visible(False)
      ax.tick_params(left = False, bottom = False)
      return plt
def plot2(df,xlabel,ylabel,actual,cnnprediction,rfprediction,knnprediction,cnnlabel,rflabel,knnlabel,title):
      plt.figure(figsize=(15, 9))
      plt.title(title,color = "black", size = 18)
      plt.xlabel(xlabel, color = "black", size = 18)
      plt.ylabel(ylabel, color = "black", size = 18)
      plt.rc('xtick', labelsize=15)
      plt.rc('ytick', labelsize=15)
      sns.lineplot(x = df["Pair Time Duration"],y= df[actual],color="black",label = actual)
      ax=sns.lineplot(x = df["Pair Time Duration"],y= df[rfprediction],color="grey",label = rflabel)
      ax=sns.lineplot(x = df["Pair Time Duration"],y= df[knnprediction],color="blue",label = knnlabel)
      ax=sns.lineplot(x = df["Pair Time Duration"],y= df[cnnprediction],color="grey",label = cnnlabel)
      ax.spines['right'].set_visible(False)
      ax.spines['top'].set_visible(False)
      ax.spines['left'].set_visible(False)
      ax.spines['bottom'].set_visible(False)
      ax.tick_params(left = False, bottom = False)
      return plt
from cProfile import label


def plot3(df,xlabel,ylabel,actual,prediction,predictionlabel,title):
      #fig1, ax = plt.subplots()
      plt.figure(figsize=(15, 9))
      plt.title(title, color = "black", size = 18)
      plt.xlabel(xlabel, color = "black", size = 18)
      plt.ylabel(ylabel, color = "black", size = 18)
      plt.xticks( range(0,60,5) )
      sns.lineplot(x = df["pair_Time_Duration"],y= df[actual],color="black", label = actual)
      ax=sns.lineplot(x = df["pair_Time_Duration"],y= df[prediction],color="b", label = predictionlabel)

      #ax.get_legend ().remove ()
      ax.spines['right'].set_visible(False)
      ax.spines['top'].set_visible(False)
      ax.spines['left'].set_visible(False)
      ax.spines['bottom'].set_visible(False)
      ax.tick_params(left = False, bottom = False)
      return plt
```
